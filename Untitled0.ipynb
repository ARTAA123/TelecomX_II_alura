## Preparación de los datos

### Extracción del Archivo Tratado

```python
import pandas as pd

datos = pd.read_csv('datos_procesados_telecom.csv')
datos.head()
```

```python
datos.info()
```

```python
datos.describe().round(2)
```

###Eliminación de columnas irrelevantes

```python
datos.drop(columns=['customerID'], inplace=True)
datos.sample(5)
```

### Verificación de la Proporción de Cancelación (Churn)

```python
datos['Churn'].value_counts(normalize=True).round(2) * 100
```

### Cambio de valores booleanos a codificación binaria

```python
datos['Genero'] = datos['Genero'].replace({'Female': 0, 'Male': 1})
datos['CiudadanoSenior'] = datos['CiudadanoSenior'].replace({False:0, True:1})
```

### Encoding de variables categoricas

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

# Separar features y target antes del encoding
X = datos.drop(columns=["Churn"])
y = datos["Churn"]

# Columnas categóricas (solo las que son texto)
categoricas = ["ServicioInternet", "TipoContrato", "MetodoPago"]

# Definir transformador
one_hot_enc = make_column_transformer(
    (OneHotEncoder(handle_unknown="ignore"), categoricas),
    remainder="passthrough"
)

# Ajustar y transformar solo X
X_transformed = one_hot_enc.fit_transform(X)

# Pasar a DataFrame
X_transformed = pd.DataFrame(X_transformed, columns=one_hot_enc.get_feature_names_out())
```

```python
# Calcular la moda de la columna
moda_seguridad = X_transformed['remainder__SeguridadEnLinea'].mode()[0]

# Rellenar los NaN con la moda
X_transformed['remainder__SeguridadEnLinea'].fillna(moda_seguridad, inplace=True)
```

```python
X_transformed.sample(5)
```

##Modelado predictivo

###Separación de datos


```python
from sklearn.model_selection import train_test_split

# X_transformed: features preprocesadas
# y: target Churn

X_train, X_test, y_train, y_test = train_test_split(
    X_transformed, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Train size: {X_train.shape}, Test size: {X_test.shape}")
```

### Balanceo de clases con SMOTE

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Revisar distribución original
print("Distribución original:", Counter(y_train))

# Crear objeto SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Verificar distribución después de SMOTE
print("Distribución después de SMOTE:", Counter(y_train_res))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)
```

### Creación de modelos

#### Regresión lineal, un modelo sensible a la escalada

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.pipeline import Pipeline

pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),  # opción MinMaxScaler si prefieres
    ('logreg', LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced'))  # balance de clases
])

pipeline_lr.fit(X_train_res_scaled, y_train_res)
y_pred_lr = pipeline_lr.predict(X_test_scaled)

print("=== Regresión Logística ===")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_lr))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred_lr))
```

#### RandomForest, modelo no sensible a escalada

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 150, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train_res_scaled, y_train_res)
best_rf = grid_search.best_estimator_
print("Mejores parámetros:", grid_search.best_params_)
```

### Evaluación de los modelos

```python
y_pred_best = best_rf.predict(X_test_scaled)
y_proba_best = best_rf.predict_proba(X_test_scaled)[:,1]

from sklearn.metrics import roc_auc_score

print("=== Random Forest Optimizado ===")
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_best))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred_best))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_best))
```

```python
importances = best_rf.feature_importances_
feature_importances = pd.DataFrame({
    'Features': X_transformed.columns,  # columnas tras One-Hot
    'Importances': (importances*100).round(2)
}).sort_values('Importances', ascending=False)

feature_importances.head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_features = feature_importances.head(15)
plt.figure(figsize=(10,6))
sns.barplot(x='Importances', y='Features', data=top_features, palette='viridis')
plt.title("Top 15 Features más importantes para Random Forest")
plt.show()
```

```python
import plotly.graph_objects as go
import plotly.express as px
from sklearn.metrics import roc_curve, auc, precision_recall_curve
```

```python
# -----------------------------------------------------
# 1️⃣ Curva ROC interactiva
# -----------------------------------------------------
fpr_lr, tpr_lr, _ = roc_curve(y_test, pipeline_lr.predict_proba(X_test_scaled)[:,1])
roc_auc_lr = auc(fpr_lr, tpr_lr)

fpr_rf, tpr_rf, _ = roc_curve(y_test, best_rf.predict_proba(X_test_scaled)[:,1])
roc_auc_rf = auc(fpr_rf, tpr_rf)

fig_roc = go.Figure()
fig_roc.add_trace(go.Scatter(x=fpr_lr, y=tpr_lr, mode='lines', name=f'Logistic Regression (AUC={roc_auc_lr:.3f})', line=dict(color='blue')))
fig_roc.add_trace(go.Scatter(x=fpr_rf, y=tpr_rf, mode='lines', name=f'Random Forest (AUC={roc_auc_rf:.3f})', line=dict(color='green')))
fig_roc.add_trace(go.Scatter(x=[0,1], y=[0,1], mode='lines', line=dict(color='gray', dash='dash'), showlegend=False))
fig_roc.update_layout(title='Curva ROC Interactiva', xaxis_title='False Positive Rate', yaxis_title='True Positive Rate', width=700, height=500)
fig_roc.show()
```

```python
# -----------------------------------------------------
# 2️⃣ Curva Precision-Recall interactiva
# -----------------------------------------------------
precision_lr, recall_lr, _ = precision_recall_curve(y_test, pipeline_lr.predict_proba(X_test_scaled)[:,1])
precision_rf, recall_rf, _ = precision_recall_curve(y_test, best_rf.predict_proba(X_test_scaled)[:,1])

fig_pr = go.Figure()
fig_pr.add_trace(go.Scatter(x=recall_lr, y=precision_lr, mode='lines', name='Logistic Regression', line=dict(color='blue')))
fig_pr.add_trace(go.Scatter(x=recall_rf, y=precision_rf, mode='lines', name='Random Forest', line=dict(color='green')))
fig_pr.update_layout(title='Curva Precision-Recall Interactiva', xaxis_title='Recall', yaxis_title='Precision', width=700, height=500)
fig_pr.show()
```

```python
# -----------------------------------------------------
# 3️⃣ Tabla de métricas comparativa
# -----------------------------------------------------
import pandas as pd
from sklearn.metrics import f1_score, recall_score, precision_score, accuracy_score

metrics_df = pd.DataFrame({
    'Modelo': ['Logistic Regression', 'Random Forest'],
    'Accuracy': [accuracy_score(y_test, y_pred_lr), accuracy_score(y_test, y_pred_best)],
    'Precision (Churn=1)': [precision_score(y_test, y_pred_lr), precision_score(y_test, y_pred_best)],
    'Recall (Churn=1)': [recall_score(y_test, y_pred_lr), recall_score(y_test, y_pred_best)],
    'F1-score (Churn=1)': [f1_score(y_test, y_pred_lr), f1_score(y_test, y_pred_best)],
    'ROC-AUC': [roc_auc_lr, roc_auc_rf]
}).round(3)

fig_table = go.Figure(data=[go.Table(
    header=dict(values=list(metrics_df.columns), fill_color='paleturquoise', align='center'),
    cells=dict(values=[metrics_df[col] for col in metrics_df.columns], fill_color='lavender', align='center'))
])
fig_table.update_layout(title='Comparativa de métricas', width=900, height=300)
fig_table.show()
```

### 1️⃣ KNN – Importancia de Variables con Permutation Importance

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import permutation_importance

# Entrenar KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_res, y_train_res)

# Evaluar en test
y_pred_knn = knn.predict(X_test)
print("Accuracy KNN:", accuracy_score(y_test, y_pred_knn))

# Permutation Importance
perm_importance = permutation_importance(knn, X_test, y_test, n_repeats=10, random_state=42)

# Crear DataFrame de importancia
knn_importances = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': perm_importance.importances_mean
}).sort_values(by='Importance', ascending=False)

knn_importances.head(15)  # Top 15 variables más influyentes
```

```python
# --- Visualización ---
plt.figure(figsize=(10,6))
top_features = knn_importances.head(15)
plt.barh(top_features['Feature'], top_features['Importance'], color="skyblue")
plt.xlabel("Importancia")
plt.ylabel("Variable")
plt.title("Top 15 Variables más Influyentes en KNN")
plt.gca().invert_yaxis()
plt.show()
```

## Análisis de las Gráficas y Tablas

Aquí tienes un análisis de las visualizaciones y tablas generadas:

1.  **Matriz de Correlación de Variables:** Este mapa de calor muestra la correlación entre todas las variables. Los colores cálidos (rojo/naranja) indican correlación positiva, mientras que los colores fríos (azul) indican correlación negativa. Podemos observar la relación entre pares de variables.

2.  **Top 10 variables más correlacionadas con Churn (Gráfico de barras):** Este gráfico resalta las 10 variables con la mayor correlación (en valor absoluto) con la variable `Churn`. Puedes ver cuáles tienen una relación más fuerte (positiva o negativa) con la cancelación del servicio.

3.  **Correlación de cada variable con Churn (Tabla):** Similar al gráfico anterior, esta tabla muestra el coeficiente de correlación de cada variable con `Churn`, ordenado de mayor a menor. Esto te da una visión completa de la fuerza y dirección de la relación de cada feature con el target.

4.  **Tiempo de contrato vs Cancelación (Boxplot):** Este gráfico de cajas muestra la distribución del tiempo de permanencia (`Permanenciadelcliente`) para los clientes que cancelaron (Churn=1) y los que no (Churn=0). Podemos observar que los clientes que cancelan tienden a tener un tiempo de permanencia menor.

5.  **Gasto total vs Cancelación (Boxplot):** Similar al anterior, este boxplot muestra la distribución del gasto total (`FacturacionTotal`) para ambos grupos de Churn. Se puede apreciar que los clientes que cancelan tienen, en promedio, un gasto total menor.

6.  **Curva ROC Interactiva:** La curva ROC (Receiver Operating Characteristic) es una métrica para evaluar el rendimiento de un modelo de clasificación. Muestra la tasa de verdaderos positivos (TPR) frente a la tasa de falsos positivos (FPR) para diferentes umbrales de clasificación. Un AUC (Área bajo la curva) más cercano a 1 indica un mejor rendimiento del modelo. En este gráfico, puedes comparar el rendimiento de Regresión Logística y Random Forest.

7.  **Curva Precision-Recall Interactiva:** La curva Precision-Recall es especialmente útil en casos de clases desbalanceadas (como el Churn). Muestra la precisión frente al recall para diferentes umbrales. Un área bajo la curva Precision-Recall más alta indica un mejor rendimiento, especialmente cuando predecir la clase positiva (Churn=1) es importante.

8.  **Tabla de métricas comparativa:** Esta tabla resume las métricas clave de evaluación (Accuracy, Precision, Recall, F1-score y ROC-AUC) para ambos modelos, facilitando la comparación de su rendimiento.

9.  **Top 15 Features Importantes - Random Forest (Gráfico de barras):** Este gráfico muestra la importancia de cada variable según el modelo Random Forest. Las variables con mayor importancia contribuyen más a la predicción de Churn.

10. **Permutation Importance - KNN:** Este análisis muestra cuánto impacta cada variable en el rendimiento del modelo KNN cuando se permuta aleatoriamente su valor. Las variables con mayor "Importance" son más influyentes en la predicción.

En resumen, estas visualizaciones te ayudan a entender las relaciones entre las variables, identificar las más relevantes para predecir Churn y comparar el desempeño de los modelos de clasificación.
