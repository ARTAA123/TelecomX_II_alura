## 📌 Introducción

El presente trabajo tiene como objetivo aplicar **técnicas de Machine Learning** para predecir la **probabilidad de churn (abandono de clientes)** en una empresa de telecomunicaciones.

Para ello se realiza:

* **Preprocesamiento de datos**: limpieza, normalización y balanceo de clases.
* **Análisis exploratorio**: visualización de correlaciones y patrones relevantes en las variables.
* **Entrenamiento de modelos supervisados**: Regresión Logística, Random Forest, SVM y otros clasificadores.
* **Evaluación de rendimiento** mediante métricas como *accuracy*, *precision*, *recall*, *F1-score* y *ROC-AUC*.
* **Interpretación de resultados**, identificando las variables más influyentes en la predicción del churn.

Este proyecto no solo busca construir un modelo predictivo, sino también proporcionar **insights valiosos** que ayuden a la toma de decisiones estratégicas para **retener clientes y mejorar la fidelización**.


## Preparación de los datos

### Extracción del Archivo Tratado

````python
import pandas as pd

datos = pd.read_csv('datos_procesados_telecom.csv')
datos.head()
````

````python
datos.info()
````

````python
datos.describe().round(2)
````

###Eliminación de columnas irrelevantes

````python
datos.drop(columns=['customerID'], inplace=True)
datos.sample(5)
````

### Verificación de la Proporción de Cancelación (Churn)

````python
datos['Churn'].value_counts(normalize=True).round(2) * 100
````

### Cambio de valores booleanos a codificación binaria

````python
datos['Genero'] = datos['Genero'].replace({'Female': 0, 'Male': 1})
datos['CiudadanoSenior'] = datos['CiudadanoSenior'].replace({False:0, True:1})
````

### Encoding de variables categoricas

````python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

# Separar features y target antes del encoding
X = datos.drop(columns=["Churn"])
y = datos["Churn"]

# Columnas categóricas (solo las que son texto)
categoricas = ["ServicioInternet", "TipoContrato", "MetodoPago"]

# Definir transformador
one_hot_enc = make_column_transformer(
    (OneHotEncoder(drop="first",handle_unknown="ignore"), categoricas),
    remainder="passthrough"
)

# Ajustar y transformar solo X
X_transformed = one_hot_enc.fit_transform(X)

# Pasar a DataFrame
X_transformed = pd.DataFrame(X_transformed, columns=one_hot_enc.get_feature_names_out())
````

````python
# Calcular la moda de la columna
moda_seguridad = X_transformed['remainder__SeguridadEnLinea'].mode()[0]

# Rellenar los NaN con la moda
X_transformed['remainder__SeguridadEnLinea'].fillna(moda_seguridad, inplace=True)
````

````python
X_transformed.sample(5)
````

````python
# =====================================================
# ⚡ ANÁLISIS DE MULTICOLINEALIDAD (VIF)
# =====================================================
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calcular_vif(df):
    vif_data = pd.DataFrame()
    vif_data["feature"] = df.columns
    vif_data["VIF"] = [
        variance_inflation_factor(df.values, i) for i in range(df.shape[1])
    ]
    return vif_data.sort_values("VIF", ascending=False)

vif_df = calcular_vif(X_transformed)

print("=== Factores de Inflación de Varianza (VIF) ===")
print(vif_df.head(15))  # mostrar top 15
````

````python
# =====================================================
# 📌 ELIMINAR VARIABLES ALTAMENTE COLINEALES (opcional)
# =====================================================
# Ejemplo: eliminar FacturaciónTotal si FacturaciónMensual ya está incluida
cols_a_eliminar = ["remainder__FacturacionTotal"]  # ajusta según VIF
X_transformed.drop(columns=cols_a_eliminar, inplace=True)
````

##Modelado predictivo

###Separación de datos


````python
from sklearn.model_selection import train_test_split

# X_transformed: features preprocesadas
# y: target Churn

X_train, X_test, y_train, y_test = train_test_split(
    X_transformed, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Train size: {X_train.shape}, Test size: {X_test.shape}")
````

### Balanceo de clases con SMOTE

````python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Revisar distribución original
print("Distribución original:", Counter(y_train))

# Crear objeto SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Verificar distribución después de SMOTE
print("Distribución después de SMOTE:", Counter(y_train_res))
````

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaler = StandardScaler()
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)
````

### Creación de modelos

#### Regresión lineal, un modelo sensible a la escalada

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.pipeline import Pipeline

pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),  # opción MinMaxScaler si prefieres
    ('logreg', LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced'))  # balance de clases
])

pipeline_lr.fit(X_train_res_scaled, y_train_res)
y_pred_lr = pipeline_lr.predict(X_test_scaled)

print("=== Regresión Logística ===")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_lr))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred_lr))
````

#### RandomForest, modelo no sensible a escalada

````python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 150, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train_res_scaled, y_train_res)
best_rf = grid_search.best_estimator_
print("Mejores parámetros:", grid_search.best_params_)
````

### Evaluación de los modelos

````python
y_pred_best = best_rf.predict(X_test_scaled)
y_proba_best = best_rf.predict_proba(X_test_scaled)[:,1]

from sklearn.metrics import roc_auc_score

print("=== Random Forest Optimizado ===")
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_best))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred_best))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_best))
````

````python
importances = best_rf.feature_importances_
feature_importances = pd.DataFrame({
    'Features': X_transformed.columns,  # columnas tras One-Hot
    'Importances': (importances*100).round(2)
}).sort_values('Importances', ascending=False)

feature_importances.head(20)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

top_features = feature_importances.head(15)
plt.figure(figsize=(10,6))
sns.barplot(x='Importances', y='Features', data=top_features, palette='viridis')
plt.title("Top 15 Features más importantes para Random Forest")
plt.show()
````

````python
import plotly.graph_objects as go
import plotly.express as px
from sklearn.metrics import roc_curve, auc, precision_recall_curve
````

````python
# -----------------------------------------------------
# 1️⃣ Curva ROC interactiva
# -----------------------------------------------------
fpr_lr, tpr_lr, _ = roc_curve(y_test, pipeline_lr.predict_proba(X_test_scaled)[:,1])
roc_auc_lr = auc(fpr_lr, tpr_lr)

fpr_rf, tpr_rf, _ = roc_curve(y_test, best_rf.predict_proba(X_test_scaled)[:,1])
roc_auc_rf = auc(fpr_rf, tpr_rf)

fig_roc = go.Figure()
fig_roc.add_trace(go.Scatter(x=fpr_lr, y=tpr_lr, mode='lines', name=f'Logistic Regression (AUC={roc_auc_lr:.3f})', line=dict(color='blue')))
fig_roc.add_trace(go.Scatter(x=fpr_rf, y=tpr_rf, mode='lines', name=f'Random Forest (AUC={roc_auc_rf:.3f})', line=dict(color='green')))
fig_roc.add_trace(go.Scatter(x=[0,1], y=[0,1], mode='lines', line=dict(color='gray', dash='dash'), showlegend=False))
fig_roc.update_layout(title='Curva ROC Interactiva', xaxis_title='False Positive Rate', yaxis_title='True Positive Rate', width=700, height=500)
fig_roc.show()
````

````python
# -----------------------------------------------------
# 2️⃣ Curva Precision-Recall interactiva
# -----------------------------------------------------
precision_lr, recall_lr, _ = precision_recall_curve(y_test, pipeline_lr.predict_proba(X_test_scaled)[:,1])
precision_rf, recall_rf, _ = precision_recall_curve(y_test, best_rf.predict_proba(X_test_scaled)[:,1])

fig_pr = go.Figure()
fig_pr.add_trace(go.Scatter(x=recall_lr, y=precision_lr, mode='lines', name='Logistic Regression', line=dict(color='blue')))
fig_pr.add_trace(go.Scatter(x=recall_rf, y=precision_rf, mode='lines', name='Random Forest', line=dict(color='green')))
fig_pr.update_layout(title='Curva Precision-Recall Interactiva', xaxis_title='Recall', yaxis_title='Precision', width=700, height=500)
fig_pr.show()
````

### 1️⃣ KNN – Importancia de Variables con Permutation Importance

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import permutation_importance

# Entrenar KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_res, y_train_res)

# Evaluar en test
y_pred_knn = knn.predict(X_test)
print("Accuracy KNN:", accuracy_score(y_test, y_pred_knn))

# Permutation Importance
perm_importance = permutation_importance(knn, X_test, y_test, n_repeats=10, random_state=42)

# Crear DataFrame de importancia
knn_importances = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': perm_importance.importances_mean
}).sort_values(by='Importance', ascending=False)

knn_importances.head(15)  # Top 15 variables más influyentes
````

Se recomienda usar Random Forest para predicción general y regresión logística para interpretabilidad.


````python
# =====================================================
# 5️⃣ GRÁFICOS INTERACTIVOS ROC Y PRECISION-RECALL
# =====================================================
from sklearn.metrics import roc_curve, precision_recall_curve
import plotly.graph_objects as go

# Función para ROC interactivo
def plot_roc(y_true, y_scores, model_name):
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=fpr, y=tpr, mode='lines', name=f'{model_name}'))
    fig.add_trace(go.Scatter(x=[0,1], y=[0,1], mode='lines', name='Random', line=dict(dash='dash')))
    fig.update_layout(
        title=f'Curva ROC - {model_name}',
        xaxis_title='False Positive Rate',
        yaxis_title='True Positive Rate',
        template='plotly_white'
    )
    fig.show()

# Función para Precision-Recall interactivo
def plot_pr(y_true, y_scores, model_name):
    precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=recall, y=precision, mode='lines', name=f'{model_name}'))
    fig.update_layout(
        title=f'Curva Precision-Recall - {model_name}',
        xaxis_title='Recall',
        yaxis_title='Precision',
        template='plotly_white'
    )
    fig.show()
````

````python
# =====================================================
# Calcular probabilidades para cada modelo
# =====================================================
# Regresión Logística
y_proba_lr = pipeline_lr.predict_proba(X_test_scaled)[:,1]

# Random Forest
y_proba_rf = best_rf.predict_proba(X_test_scaled)[:,1]

# KNN
y_proba_knn = knn.predict_proba(X_test_scaled)[:,1]

# Graficar ROC interactivo
plot_roc(y_test, y_proba_lr, "Regresión Logística")
plot_roc(y_test, y_proba_rf, "Random Forest")
plot_roc(y_test, y_proba_knn, "KNN")

# Graficar Precision-Recall interactivo
plot_pr(y_test, y_proba_lr, "Regresión Logística")
plot_pr(y_test, y_proba_rf, "Random Forest")
plot_pr(y_test, y_proba_knn, "KNN")
````

**Explicación:**  
- **Curva ROC:** Muestra la capacidad del modelo para distinguir entre clientes que cancelan y no cancelan.  
  - Más cerca de la esquina superior izquierda = mejor rendimiento.  
- **Curva Precision-Recall:** Útil cuando la clase positiva (Churn=1) es minoritaria.
